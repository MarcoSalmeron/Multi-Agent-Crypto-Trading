# Contenido
El conjunto de datos contiene un archivo CSV para cada criptomoneda. La historia de precios está disponible de manera diaria desde el 28 de abril de 2013. Este dataset incluye la información histórica de precios de algunas de las principales criptomonedas por capitalización de mercado.

- **Date**: fecha de observación
- **Open**: precio de apertura en el día dado
- **High**: precio más alto en el día dado
- **Low**: precio más bajo en el día dado
- **Close**: precio de cierre en el día dado
- **Volume**: volumen de transacciones en el día dado
- **Market Cap**: capitalización de mercado en USD

**Agradecimientos**
Estos datos provienen de *coinmarketcap* y son de uso libre.

**Inspiración**
Algunas de las preguntas que se podrían inferir de este dataset son:

- ¿Cómo cambiaron los precios históricos / capitalizaciones de mercado de varias monedas a lo largo del tiempo?
- Predicción del precio futuro de las criptomonedas
- ¿Qué monedas son más volátiles y cuáles más estables?
- ¿Cómo se correlacionan las fluctuaciones de precios entre distintas monedas?
- Tendencias estacionales en las fluctuaciones de precios

# 2-) Construcción del modelo de Regresión Lineal

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
```

## 2.1-) Auxiliares

```python
def train_val_test_split(df, rstate=42, shuffle=True, stratify=None):
    strat = df[stratify] if stratify else None
    train_set, test_set = train_test_split(
        df, test_size=0.4, random_state=rstate, shuffle=shuffle, stratify=strat)
    strat = test_set[stratify] if stratify else None
    val_set, test_set = train_test_split(
        test_set, test_size=0.5, random_state=rstate, shuffle=shuffle, stratify=strat)
    return (train_set, val_set, test_set)
```

```python
def remove_labels(df, label_name):
    X = df.drop(label_name, axis=1)
    y = df[label_name].copy()
    return (X, y)
```

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", RobustScaler())
])
```

```python
class CustomOneHotEncoder(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._oh = OneHotEncoder()
        self._columns = None

    def fit(self, X, y=None):
        X_cat = X.select_dtypes(include=['object'])
        self._columns = pd.get_dummies(X_cat).columns
        self._oh.fit(X_cat)
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_cat = X_copy.select_dtypes(include=['object'])
        X_num = X_copy.select_dtypes(exclude=['object'])
        X_cat_oh = self._oh.transform(X_cat)
        X_cat_oh = pd.DataFrame(X_cat_oh.toarray(),
                                columns=self._columns,
                                index=X_copy.index)
        X_copy.drop(list(X_cat), axis=1, inplace=True)
        return X_copy.join(X_cat_oh)
```

```python
class DataFramePreparer(BaseEstimator, TransformerMixin):

    def __init__(self):
        self._full_pipeline = None
        self._columns = None

    def fit(self, X, y=None):
        num_attribs = list(X.select_dtypes(exclude=['object']))
        cat_attribs = list(X.select_dtypes(include=['object']))
        self._full_pipeline = ColumnTransformer([
                ("num", num_pipeline, num_attribs),
                ("cat", CustomOneHotEncoder(), cat_attribs),
        ])
        self._full_pipeline.fit(X)
        self._columns = pd.get_dummies(X).columns
        return self

    def transform(self, X, y=None):
        X_copy = X.copy()
        X_prep = self._full_pipeline.transform(X_copy)
        return pd.DataFrame(X_prep,
                            columns=self._columns,
                            index=X_copy.index)
```

## 2.2-) Dividir el conjunto de datos

```python
# tirar columnas sin relevancia predictiva
columnas_irrelevantes = ['SNo', 'Name', 'Symbol', 'Date']

Ethereum.drop(columns=columnas_irrelevantes, inplace=True)

Bitcoin.drop(columns=columnas_irrelevantes, inplace=True)

Dogecoin.drop(columns=columnas_irrelevantes, inplace=True)
```

```python
train_set_ethereum, val_set_ethereum, test_set_ethereum = train_val_test_split(Ethereum)

train_set_bitcoin, val_set_bitcoin, test_set_bitcoin = train_val_test_split(Bitcoin)

train_set_doge, val_set_doge, test_set_doge = train_val_test_split(Dogecoin)
```

```python
target = 'Prediction'
```

```python
X_train_Ethereum, y_train_Ethereum = remove_labels(train_set_ethereum, target)
X_val_Ethereum, y_val_Ethereum = remove_labels(val_set_ethereum, target)
X_test_Ethereum, y_test_Ethereum = remove_labels(test_set_ethereum, target)
```

```python
X_train_Bitcoin, y_train_Bitcoin = remove_labels(train_set_bitcoin, target)
X_val_Bitcoin, y_val_Bitcoin = remove_labels(val_set_bitcoin, target)
X_test_Bitcoin, y_test_Bitcoin = remove_labels(test_set_bitcoin, target)
```

```python
X_train_Dogecoin, y_train_Dogecoin = remove_labels(train_set_doge, target)
X_val_Dogecoin, y_val_Dogecoin = remove_labels(val_set_doge, target)
X_test_Dogecoin, y_test_Dogecoin = remove_labels(test_set_doge, target)
```

## 2.3-) Entrenamiento del modelo

```python
pipeline_Ethereum = Pipeline([
    ("prep", DataFramePreparer()),
    ("model", LinearRegression()),
])

pipeline_Bitcoin = Pipeline([
    ("prep", DataFramePreparer()),
    ("model", LinearRegression()),
])

pipeline_Dogecoin = Pipeline([
    ("prep", DataFramePreparer()),
    ("model", LinearRegression()),
])
```

```python
pipeline_Ethereum.fit(X_train_Ethereum, y_train_Ethereum)
```

```python
pipeline_Bitcoin.fit(X_train_Bitcoin, y_train_Bitcoin)
```

```python
pipeline_Dogecoin.fit(X_train_Dogecoin, y_train_Dogecoin)
```

## 2.4-) Evaluación del modelo
